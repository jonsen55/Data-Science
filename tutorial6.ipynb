### 1. Introduction to Matplotlib

##### What is Matplotlib?

Matplotlib is a Python package for 2D plotting that generates production-quality graphs. It supports interactive and non-interactive plotting, and can save images in several output formats (PNG, PS, and others). It can use multiple window toolkits (GTK+, wxWidgets, Qt, and so on) and it provides a wide variety of plot types.

##### Write a Python script to plot a simple line graph.

```python
import matplotlib.pyplot as plt
x = [1,2,5,7,9]
y = [6,7,8,9,10]
plt.plot(x,y)
```

##### Create a figure with two subplots one showing a sine wave and the other a cosine wave.

```python
import numpy as np
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# First plot
axes[0].plot(x, y1)

# Second plot
axes[1].plot(x, y2)
```
